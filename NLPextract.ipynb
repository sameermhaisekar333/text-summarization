```python
def create_directory(path,filename,content):
    try:
        file_path = path.joinpath(filename)
        
        if not os.path.exists(path):
            os.mkdir(path)
            creat_file(file_path,content)
            print("{} created ".format(file_path))
        else:
            creat_file(file_path,content)
            print("{} created ".format(file_path))
    except Exception as e :
        print("Error at creation of directory",e)
        
def creat_file(name,content):
    with open(name,"w") as f:
        f.write(content)
```

```python
# read stopword folder files

def read_stopwords_file(path):
    with open(path,"rb")as fp:
        content =fp.readlines()
#         print(content)
#         s = bytes.decode(content, 'latin-1')
        res = list(map(lambda x:x.decode('latin-1').strip().lower(),content))
#         print(res)
#         print(path,type(path))            
        if 'Currencies'in str(path): 
#             print(True)
            res_c = list(map(lambda x:x.split("|")[0].strip(),res))
            return res_c
        if 'DatesandNumbers'in str(path):
#             print(True)
            res_d = list(map(lambda x:x.split("|")[0].strip(),res))
            return res_d
        if 'Geographic'in str(path):
#             print(True)
            res_g = list(map(lambda x:x.split("|")[0].strip(),res))
            return res_g
        if 'Names' in str(path):
#             print(True)
            res_n = list(map(lambda x:x.split("|")[0].strip(),res))
            return res_n
       
        else: 
#             print(False)
                pass
        return res
```

```python
def getwords_from_stopwords_files(stopwordpath):
    allstopwords = []
    files = [Path(x) for x in os.listdir(stopwordpath)]
#     print(files_name)
    for file in files:
        file_path = stopwordpath.joinpath(file)
        print(file_path)
        words = read_stopwords_file(file_path)
        allstopwords.extend(words)
    return list(set(allstopwords))
```

```python
def read_text_id_files(path):
    with open(path,"r") as fp:
        c = fp.read()
        return c 
```

```python
def compare_words_of_list(from_list,find_in_list,common):
    word_list =[]
    for word in from_list:
        if common:
            if (word in find_in_list):
                    word_list.append(word)
        else:
            if not (word in find_in_list):
                word_list.append(word)
            
                
    return dict(words=word_list,length= len(word_list))
```

```python
# Analysis of Readability
def break_sentences(text):
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)
    return list(doc.sents)

# Returns Number of Words in the text
def word_count(text):
    sentences = break_sentences(text)
    words = 0 
    for sentence in sentences:
        words += len([token for token in sentence])
    return words
# Returns the number of sentences in the text

def sentence_count(text):
    sentences = break_sentences(text)
    return len(sentences)

def syllables_count(word):   
    word=word.lower()
    return textstatistics().syllable_count(word)

def avg_sentence_length(text):
    words = word_count(text)
    sentences = sentence_count(text)
    average_sentence_length = float(words / sentences)
    return average_sentence_length

# Return total Difficult Words in a text
def difficult_words(text):
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)
    # Find all words in the text
    words = []
    sentences = break_sentences(text)
    for sentence in sentences:
        words += [str(token) for token in sentence]

    # difficult words are those with syllables >= 2
    # easy_word_set is provide by Textstat as
    # a list of common words
    diff_words_set = set()
    
    for word in words:
        syllable_count = syllables_count(word)
        if word not in nlp.Defaults.stop_words and syllable_count >= 2:
            diff_words_set.add(word)
#     print( len(diff_words_set))
    return len(diff_words_set)
def gunning_fog(text):
    per_diff_words = (difficult_words(text) / word_count(text) * 100) + 5
    grade = 0.4 * (avg_sentence_length(text) + per_diff_words)
    return grade
```

```python
vowels = ['a', 'e', 'i', 'o', 'u']
def countsyllble(textlist,vowels):
    wlist = []
    clist =[]
    for word in textlist:
       
        count = 0
        ends =('ed','es')
    #     print(word)
        for ch in word:
            if ch in vowels:
                count +=1
        if word.endswith(ends):
            count -=1
        wlist.append(word)
        clist.append(count)
    
    return dict(words = wlist,count= clist)
```

```python
def countOccurrences(text,wclist):
    wordl=[]
    countl=[]
    for w in wclist:
        # Get the regex to be checked
        regexPattern = "\b(" + w + ")\b"

        # Variable to count total
        # occurrences of the given word
        count=0

        for m in re.finditer(regexPattern, text):
            # Increment count
            count=count+1

        # Print the occurrences of the word
        wordl.append(w)
        countl.append(count)
    return  dict(words = wordl,count= countl)
```

```python
def avgwordlenght(text):
    sentences = break_sentences(text)
    words =[]
    char =[]
    for sentence in sentences:
        for word in sentence:
            words.append(word)
    
    for ch in words:
        char.append(len(ch))
        
#     print(sum(char) )
    avglength = sum(char)/len(words)    
    return avglength
```

```python

def creating_output_for_text_id(x):
    try :
#         print(x,type(x),"->",x['fileName']) 
        fileidpath = filedirpath.joinpath(x['fileName'])
        print("reading file path",fileidpath)
        fileidtextcontent = read_text_id_files(fileidpath)
        fileidtext = fileidtextcontent.lower()
        print("filecontent length ",len(fileidtext),type(fileidtext))
        tokensfileid = word_tokenize(fileidtext)
        print("tokenized file length ",len(tokensfileid),type(tokensfileid))
        cleanedidwords = compare_words_of_list(from_list=tokensfileid,find_in_list=allstopwords,common=False)
        print("length of cleanedidwords ",cleanedidwords['length'])

        #POSITIVE SCORE
        positivewordsid = compare_words_of_list(from_list=cleanedidwords['words'],find_in_list=postiveword_tokens,common=True)

        print("positive score",positivewordsid['length'])
    #     pirnt(positivewordsid['words'])
        positive_score = positivewordsid['length']

        #NEGATIVE SCORE
        negativewordsid = compare_words_of_list(from_list=cleanedidwords['words'],find_in_list=negativeword_tokens,common=True)
        print("negative score",negativewordsid['length'])
        negative_score =  negativewordsid['length']
    #     pirnt(negativewordsid['words'])

        #POLARITY SCORE
        polarity_score = (positivewordsid['length'] - negativewordsid['length']) / ((positivewordsid['length'] + negativewordsid['length'])+0.000001)
        print("polarity_score",polarity_score)

        #SUBJECTIVITY SCORE
        subjectivity_score= (positivewordsid['length'] + negativewordsid['length'])/(len(tokensfileid) + 0.000001)
        print("subjectivity_score",subjectivity_score)

        #AVG SENTENCE LENGTH
        avg_sentence_length_score = avg_sentence_length(fileidtext)
        print("avg_sentence_length",avg_sentence_length_score)

        #PERCENTAGE OF COMPLEX WORDS
        total_words_score = word_count(fileidtext)
        complex_words_score = difficult_words(fileidtext)
        per_complex_words_score = (complex_words_score *100)/total_words_score
        print("percentage of complex words",per_complex_words_score)

        #FOG INDEX
        fog_score = gunning_fog(fileidtext)
        print("fog score",fog_score)

        #AVG NUMBER OF WORDS PER SENTENCE
        avg_num_word_per_sentence_score = avg_sentence_length(fileidtext)
        print("AVG NUMBER OF WORDS PER SENTENCE",avg_num_word_per_sentence_score)

        #COMPLEX WORD COUNT
        print("COMPLEX WORD COUNT ",complex_words_score)

        #WORD COUNT
        print("WORD COUNT",total_words_score)


        tokenizer = RegexpTokenizer(r'\w+')
        fileidtext_remove_punc_ = tokenizer.tokenize(fileidtext)
    #     print(type(fileidtext_remove_punc_), len(fileidtext_remove_punc_))
        stop = set(stopwords.words('english'))
        fileidtext_remove_punc_stopwords = compare_words_of_list(from_list=fileidtext_remove_punc_,find_in_list=list(stop),
                                                            common=False)
        print("after removing punctuation,stop words length",fileidtext_remove_punc_stopwords['length'])

        vowels = ['a', 'e', 'i', 'o', 'u']
        word_syllable = countsyllble(fileidtext_remove_punc_stopwords['words'],vowels=vowels)
        print("length  wordsyllable",len(word_syllable['words']))

        #SYLLABLE PER WORD
        syllble_perword_score = sum(word_syllable['count'])/len(word_syllable['words'])
        print("SYLLABLE PER WORD ",syllble_perword_score)

        #PERSONAL PRONOUNS
        prnounslist = ['I','we', 'my', 'ours','us']
        pronoun_result = countOccurrences(text=fileidtextcontent,wclist=prnounslist)
        pronoun_score= sum(pronoun_result['count'])
        print("PERSONAL PRONOUNS",pronoun_score)

        #AVG WORD LENGTH
        avg_word_length_score=avgwordlenght(fileidtext)
        print("avg_word_length_score",avg_word_length_score)

        print("###### completed ###### ")

        return pd.Series([positive_score,negative_score ,polarity_score,subjectivity_score,
                          avg_sentence_length_score,per_complex_words_score,fog_score,
                          avg_num_word_per_sentence_score,complex_words_score,total_words_score,
                          syllble_perword_score,pronoun_score,avg_word_length_score
                         ], 
                         index=['POSITIVE SCORE', 'NEGATIVE SCORE','POLARITY SCORE','SUBJECTIVITY SCORE',
                                'AVG SENTENCE LENGTH','PERCENTAGE OF COMPLEX WORDS','FOG INDEX',
                                'AVG NUMBER OF WORDS PER SENTENCE','COMPLEX WORD COUNT','WORD COUNT',
                                'SYLLABLE PER WORD','PERSONAL PRONOUNS','AVG WORD LENGTH',
                             ])
    except Exception as e:
            print("some thing weng wrong while reusult calculation ",e)

```
